```python
def parse_goodPrice(item):

    return {
        "goodInspectDay":find_item(item,"goodInspectDay"),
        "entpId":find_item(item,"entpId"),
        "goodId":find_item(item,"goodId"),
        "goodPrice":find_item(item,"goodPrice"),
        "plusoneYn":find_item(item,"plusoneYn"),
        "goodDcYn":find_item(item,"goodDcYn"),
        "inputDttm":find_item(item,"inputDttm"),
        "goodDcStartDay":find_item(item,"goodDcStartDay"),
        "goodDcEndDay":find_item(item,"goodDcEndDay"),


    }

# "상품_조사_일":find_item(item,"goodInspectDay"),
#         "업체_아이디":find_item(item,"entpId"),
#         "상품_아이디":find_item(item,"goodId"),
#         "상품_가격":find_item(item,"goodPrice"),
#         "원플러스원_여부":find_item(item,"plusoneYn"),
#         "상품_할인_여부":find_item(item,"goodDcYn"),
#         "입력_일시":find_item(item,"inputDttm"),
#         "상품_할인_시작일":find_item(item,"goodDcStartDay"),
#         "상품_할인_종료일":find_item(item,"goodDcEndDay"),
```

```python
def date_list(date):
    date_li = [date]
    for i in range(2):
        date = (datetime.datetime.strptime(date, '%Y%m%d').date()- datetime.timedelta(days = 14)).strftime('%Y%m%d')
        date_li.append(date)
    return date_li
```

```python
# date : 가장 가까운 금요일인데 데이터가 없는 날 존재. 가장 최근 날짜를 불러오는 알고리즘 필요
# goodId : 상품 Id
# 그 밖에 업체 아이디를 기준으로 판매목록을 확인할 수 있으나 현재 프로젝트에서는 불필요하다 생각됨
def price_by_date_Id(date,goodId):
    row = []
    for d in date_list(date):
        url = 'http://openapi.price.go.kr/openApiImpl/ProductPriceInfoService/getProductPriceInfoSvc.do'
        params ={'goodInspectDay' : d,
            'serviceKey' : 'p5ddB5OtzIHdNhKoNLzEv5CqH+b6xaHvHSkw6Pljww4cKNqUah/rLoOx8udwEtTeojSyi+MMYX47YPH4SeC8cg==',
            'goodId': goodId}

        response = requests.get(url, params=params)
        soup = BeautifulSoup(response.text,'lxml-xml')
        items = soup.find_all("iros.openapi.service.vo.goodPriceVO")

        for item in items:
            row.append(parse_goodPrice(item))
        
    return row
```
